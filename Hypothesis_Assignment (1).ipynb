```python
import pandas as pd
import scipy
from scipy import stats
labtat_df=pd.read_csv('LabTAT.csv')
labtat_df
```

```python
# Ho= All means are equal of all 4 laboratory TAT is equal of all 4 laboratory
# Ha= Atleast one mean is different
```

```python
stats.f_oneway(labtat_df.iloc[:,0],labtat_df.iloc[:,1],labtat_df.iloc[:,2],labtat_df.iloc[:,3])
```

```python
# p value is 2.115 greater than 0.05 means accept Null Hypothesis
# The assumption is there is no difference in Turn around time(TAT) of all 4 laboratories.
```
